```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("data.csv", delimiter=",")
data
```

```python
def create_plots(data):
    data = pd.to_numeric(data, errors='coerce').dropna()
    data = data[data < 20]
    global_data = data
    var = np.var(data)
    plt.figure(figsize=(16, 10), dpi=120)
    plt.title("График плотности ")
    sns.kdeplot(data, fill=True, color="dodgerblue", label="-")
    print("Количество данных: " + str(len(data)))
    var = np.var(data)
    print("Дисперсия: " + str(var))
    print("Среднеквадратичное отклонение: " + str(math.sqrt(var)))
    print("Математическое ожидание: " + str(np.average(data)))
    plt.legend()
    plt.figure(figsize=(16, 10), dpi=120)
    data.plot().legend()
```

```python
data1 = data.y_first.apply(lambda x: str(x).replace(",", "."))
create_plots(data1)
```

```python
data1 = data.y_second.apply(lambda x: str(x).replace(",", "."))
create_plots(data1)
```
